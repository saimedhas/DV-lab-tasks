```
import pandas as pd
import random
from datetime import datetime, timedelta

# 2. Define lists for sources, destinations, and classes
sources = ['New York', 'London', 'Paris', 'Dubai', 'Tokyo', 'Sydney', 'Mumbai', 'Singapore']
destinations = ['New York', 'London', 'Paris', 'Dubai', 'Tokyo', 'Sydney', 'Mumbai', 'Singapore']
classes = ['Economy', 'Business', 'First Class']

# 3. Create an empty list to store the generated booking records
booking_records = []

# 4. Loop 50 times to generate each booking record
for i in range(1, 51):
    # a. Assign a unique Booking_ID
    booking_id = f'BK{i:03d}'

    # b. Randomly select a Source and Destination, ensuring they are not the same
    source = random.choice(sources)
    destination = random.choice([d for d in destinations if d != source])

    # c. Generate a Flight_ID
    flight_id = f'FL{i:03d}'

    # d. Randomly select a Class
    flight_class = random.choice(classes)

    # e. Generate a random Amount for the booking
    amount = round(random.uniform(5000, 50000), 2)

    # f. Generate a random Booking_Date within a reasonable past range (e.g., last 365 days)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)
    booking_date = start_date + (end_date - start_date) * random.random()

    # g. Generate a Travelled_Date by adding a random number of days (e.g., 7 to 90 days) to the Booking_Date
    travel_days = random.randint(7, 90)
    travelled_date = booking_date + timedelta(days=travel_days)

    # h. Append the generated record as a dictionary to your list
    booking_records.append({
        'Booking_ID': booking_id,
        'Source': source,
        'Destination': destination,
        'Flight_ID': flight_id,
        'Class': flight_class,
        'Amount': amount,
        'Booking_Date': booking_date.strftime('%Y-%m-%d'),
        'Travelled_Date': travelled_date.strftime('%Y-%m-%d')
    })

# 5. Convert the list of records into a pandas DataFrame
df_airline_bookings = pd.DataFrame(booking_records)

# 6. Save the DataFrame to a CSV file named 'airline_booking_records.csv' without including the index
df_airline_bookings.to_csv('airline_booking_records.csv', index=False)

print("Generated 'airline_booking_records.csv' with 50 records.")
print("First 5 rows of the generated DataFrame:")
print(df_airline_bookings.head())
```

```
import pandas as pd

df_employees = pd.read_csv('/content/sample_data/employee_data.csv')

print("First 5 rows of df_employees:")
print(df_employees.head())
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Create a cross-tabulation of Gender and Promotion_Status
gender_promotion_crosstab = pd.crosstab(df_employees['Gender'], df_employees['Promotion_Status'])

# Create a stacked bar chart
plt.figure(figsize=(8, 6))
gender_promotion_crosstab.plot(kind='bar', stacked=True, colormap='Paired', figsize=(8, 6))
plt.title('Promotion Status by Gender (Stacked Bar Chart)')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Promotion_Status')
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Create a cross-tabulation of Gender and Promotion_Status
gender_promotion_crosstab = pd.crosstab(df_employees['Gender'], df_employees['Promotion_Status'])

# Create a grouped bar chart
plt.figure(figsize=(8, 6))
gender_promotion_crosstab.plot(kind='bar', colormap='Paired', figsize=(8, 6))
plt.title('Promotion Status by Gender (Grouped Bar Chart)')
plt.xlabel('Gender')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.legend(title='Promotion_Status')
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import seaborn as sns

# Create a cross-tabulation of Department and Job_Level
department_job_level_crosstab = pd.crosstab(df_employees['Department'], df_employees['Job_Level'])

# Normalize the data to show proportions for a segmented bar chart
department_job_level_crosstab_norm = department_job_level_crosstab.div(department_job_level_crosstab.sum(1).astype(float), axis=0)

# Create a segmented (stacked) bar chart for Department vs. Job_Level
plt.figure(figsize=(10, 7))
department_job_level_crosstab_norm.plot(kind='bar', stacked=True, colormap='Spectral', figsize=(10, 7))
plt.title('Job Level Distribution by Department (Segmented Bar Chart)')
plt.xlabel('Department')
plt.ylabel('Proportion of Employees')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Job_Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

# Create a mosaic plot for Education_Level vs. Promotion_Status
plt.figure(figsize=(10, 7))
mosaic(df_employees, ['Education_Level', 'Promotion_Status'], title='Education Level vs. Promotion Status (Mosaic Plot)')
plt.ylabel('Promotion Status')
plt.xlabel('Education Level')
plt.tight_layout()
plt.show()
```
