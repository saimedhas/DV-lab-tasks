```
import pandas as pd
import random
from datetime import datetime, timedelta

# 2. Define lists for sources, destinations, and classes
sources = ['New York', 'London', 'Paris', 'Dubai', 'Tokyo', 'Sydney', 'Mumbai', 'Singapore']
destinations = ['New York', 'London', 'Paris', 'Dubai', 'Tokyo', 'Sydney', 'Mumbai', 'Singapore']
classes = ['Economy', 'Business', 'First Class']

# 3. Create an empty list to store the generated booking records
booking_records = []

# 4. Loop 50 times to generate each booking record
for i in range(1, 51):
    # a. Assign a unique Booking_ID
    booking_id = f'BK{i:03d}'

    # b. Randomly select a Source and Destination, ensuring they are not the same
    source = random.choice(sources)
    destination = random.choice([d for d in destinations if d != source])

    # c. Generate a Flight_ID
    flight_id = f'FL{i:03d}'

    # d. Randomly select a Class
    flight_class = random.choice(classes)

    # e. Generate a random Amount for the booking
    amount = round(random.uniform(5000, 50000), 2)

    # f. Generate a random Booking_Date within a reasonable past range (e.g., last 365 days)
    end_date = datetime.now()
    start_date = end_date - timedelta(days=365)
    booking_date = start_date + (end_date - start_date) * random.random()

    # g. Generate a Travelled_Date by adding a random number of days (e.g., 7 to 90 days) to the Booking_Date
    travel_days = random.randint(7, 90)
    travelled_date = booking_date + timedelta(days=travel_days)

    # h. Append the generated record as a dictionary to your list
    booking_records.append({
        'Booking_ID': booking_id,
        'Source': source,
        'Destination': destination,
        'Flight_ID': flight_id,
        'Class': flight_class,
        'Amount': amount,
        'Booking_Date': booking_date.strftime('%Y-%m-%d'),
        'Travelled_Date': travelled_date.strftime('%Y-%m-%d')
    })

# 5. Convert the list of records into a pandas DataFrame
df_airline_bookings = pd.DataFrame(booking_records)

# 6. Save the DataFrame to a CSV file named 'airline_booking_records.csv' without including the index
df_airline_bookings.to_csv('airline_booking_records.csv', index=False)

print("Generated 'airline_booking_records.csv' with 50 records.")
print("First 5 rows of the generated DataFrame:")
print(df_airline_bookings.head())
```

```
print("\n--- Method 1: Standard Python File Reading (open().read().splitlines()) ---")
with open('airline_booking_records.csv', 'r') as file:
    content = file.read()
    lines = content.splitlines()
    for i, line in enumerate(lines[:5]):
        print(line)
```

```
import csv

print(
    "\n--- Method 2: CSV Module (csv.reader()) ---"
)
with open('airline_booking_records.csv', 'r') as file:
    csv_reader = csv.reader(file)
    for i, row in enumerate(csv_reader):
        if i < 5:
            print(row)
        else:
            break
```

```
import numpy as np

print(
    "\n--- Method 3: NumPy (numpy.loadtxt()) ---"
)
# Use dtype=str to prevent conversion errors for mixed types
# Use delimiter=',' to specify comma separation
# Use skiprows=1 to skip the header row
data_loadtxt = np.loadtxt('airline_booking_records.csv', delimiter=',', dtype=str, skiprows=1)
for row in data_loadtxt[:5]:
    print(row)
```

```
print(
    "\n--- Method 4: NumPy (numpy.genfromtxt()) ---"
)
# Use dtype=str to prevent conversion errors for mixed types
# Use delimiter=',' to specify comma separation
# Use skip_header=1 to skip the header row
data_genfromtxt = np.genfromtxt('airline_booking_records.csv', delimiter=',', dtype=str, skip_header=1)
for row in data_genfromtxt[:5]:
    print(row)
```

```
import pandas as pd

print(
    "\n--- Method 5: Pandas (pd.read_csv()) ---"
)
df_pandas = pd.read_csv('airline_booking_records.csv')
print(df_pandas.head())
```

```
print("\n--- Details for Flight_ID 'FL001' ---")
print(df_pandas[df_pandas['Flight_ID'] == 'FL001'])

print("\n--- Number of unique flight classes ---")
print(df_pandas['Class'].nunique())

print("\n--- Number of unique destinations ---")
print(df_pandas['Destination'].nunique())

print("\n--- Summary statistics for 'Amount' ---")
print(df_pandas['Amount'].describe())
```

```
print("--- Flight_IDs where Amount > ₹15,000 ---")
flight_ids_greater_than_15k = df_pandas[df_pandas['Amount'] > 15000]['Flight_ID']
print(flight_ids_greater_than_15k)
```

```
print(
    "--- Booking_IDs where Amount is between ₹20,000 and ₹40,000 ---"
)
booking_ids_between_20k_and_40k = df_pandas[
    (df_pandas['Amount'] >= 20000) & (df_pandas['Amount'] <= 40000)
]['Booking_ID']
print(booking_ids_between_20k_and_40k)
```

```
df_pandas['Booking_Date'] = pd.to_datetime(df_pandas['Booking_Date'])
df_pandas['Travelled_Date'] = pd.to_datetime(df_pandas['Travelled_Date'])
df_pandas['Booking_Lead_Time'] = df_pandas['Travelled_Date'] - df_pandas['Booking_Date']

print("DataFrame after date conversion and 'Booking_Lead_Time' calculation:")
print(df_pandas.head())
```
